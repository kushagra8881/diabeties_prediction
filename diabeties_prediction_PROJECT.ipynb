MAKING ZEROS TO THEIR RESPECTIVE MEDIAN

```python
dataset["Glucose"]=dataset["Glucose"].replace(0,dataset["Glucose"].median())
dataset["BloodPressure"]=dataset["BloodPressure"].replace(0,dataset["BloodPressure"].median())
dataset["BMI"]=dataset["BMI"].replace(0,dataset["BMI"].median())
dataset["SkinThickness"]=dataset["SkinThickness"].replace(0,dataset["SkinThickness"].median())
dataset["Insulin"]=dataset["Insulin"].replace(0,dataset["Insulin"].median())
```

```python
dataset
```

OUR MACHINE WORKING START

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
y.shape
```

```python
y=y.reshape(768,-1)
```

```python
y
```

DISTRIBUTING IN TRAIN AND TEST

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=100)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

OUR MODEL WORKING ALGOES AND THIER ACUURACY

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
train_accuracy=[]
test_accuracy=[]
for n in range(1,11):
    knn=KNeighborsClassifier(n_neighbors=n)
    knn.fit(x_train,y_train)
    train_accuracy.append(knn.score(x_train,y_train))
    test_accuracy.append(knn.score(x_test,y_test))
print(train_accuracy)
print(test_accuracy)
```

```python

plt.plot(range(1,11),test_accuracy,label="test_accurcy")
plt.plot(range(1,11),train_accuracy,label="train_accuracy")
plt.show()
```

```python
knn=KNeighborsClassifier(n_neighbors=9)
knn.fit(x_train,y_train)
print(knn.score(x_train,y_train)*100)
print(knn.score(x_test,y_test)*100)
```

```python
from sklearn.svm import SVC
classifier=SVC()
classifier.fit(x_train,y_train)
print(classifier.score(x_train,y_train)*100)
print(classifier.score(x_test,y_test)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
train_accuracy=[]
test_accuracy=[]
for n in range(1,11):
    knn=RandomForestClassifier(n_estimators=n , criterion = 'entropy')
    knn.fit(x_train,y_train)
    train_accuracy.append(knn.score(x_train,y_train))
    test_accuracy.append(knn.score(x_test,y_test))
print(train_accuracy)
print(test_accuracy)

plt.plot(range(1,11),test_accuracy,label="test_accurcy")
plt.plot(range(1,11),train_accuracy,label="train_accuracy")
plt.show()
```

```python
knn=RandomForestClassifier(n_estimators=11 , criterion = 'entropy')
knn.fit(x_train,y_train)
print(knn.score(x_train,y_train)*100)
print(knn.score(x_test,y_test)*100)
```

FEATURE SCALING AND REDOING MODEL PREDICTION

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```
